```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
from scipy.stats import chi2
```

# Random Variable

```python
# Sample size
n = 14

# Significance
alfa = 0.05
```

```python
# Generates samples from the Chi square distribution
rv = chi2.rvs(n-1, size = 10000)
```

```python
# Visualize distribution
df = pd.DataFrame(data = {"rv": rv})
sns.distplot(df["rv"], bins = 50, kde=True)
```

# Quantiles

```python
quantile = 0.1
np.round(chi2.ppf(quantile, n),2)
```

```python
quantile = 0.9
np.round(chi2.ppf(quantile, n),2)
```

# Centrality

```python
chi2.mean(n)
```

```python
chi2.var(n)
```

# PDF

```python
# Get the quantiles of 1% and 99% of probability
x = np.linspace(chi2.ppf(0.01, n), chi2.ppf(0.99, n), 100)
```

```python
# Get the value of the density
y = chi2.pdf(x, n)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, y,'--k', lw=3, alpha=0.6, label='chi2 pdf')
```
